```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import torch
import cv2
```

```python
train_df = pd.read_csv("label/train.csv")
val_df = pd.read_csv("label/val.csv")
test_df = pd.read_csv("label/test.csv")


```

```python
import numpy as np
from sklearn.metrics import f1_score


thresholds = np.arange(0.0, 1.01, 0.0001)

best_threshold = 0
best_f1 = 0


for t in thresholds:
    preds_t = (val_df['preds'] >= t).astype(int)
    f1 = f1_score(val_df['cancer'], preds_t)
    
    if f1 > best_f1:
        best_f1 = f1
        best_threshold = t

print(f"Best threshold: {best_threshold:.2f}, Best F1 on val: {best_f1:.4f}")


test_preds_t = (test_df['preds'] >= best_threshold).astype(int)
test_f1 = f1_score(test_df['cancer'], test_preds_t)

print(f"Final F1 score on test: {test_f1:.4f}")
```

```python
import numpy as np
from sklearn.metrics import f1_score
from tqdm import tqdm

thresholds = np.arange(0.0, 1.01, 0.0001)

best_threshold = 0
best_f1 = 0


for t in tqdm(thresholds):
    preds_t = (val_df['preds'] >= t).astype(int)
    f1 = f1_score(val_df['cancer'], preds_t)
    
    if f1 > best_f1:
        best_f1 = f1
        best_threshold = t

print(f"Best threshold: {best_threshold:.5f}, Best F1 on val: {best_f1:.5f}")


test_preds_t = (test_df['preds'] >= best_threshold).astype(int)
test_f1 = f1_score(test_df['cancer'], test_preds_t)

print(f"Final F1 score on test: {test_f1:.5f}")
```
